# Synthetic Features

```python
synthetic_features = df.copy()
```

```python
synthetic_features
```

```python
synthetic_features = df.copy()
synthetic_features['Fecha-I'] =  pd.to_datetime(synthetic_features['Fecha-I'], format='%Y-%m-%d %H:%M:%S')
synthetic_features['Fecha-O'] =  pd.to_datetime(synthetic_features['Fecha-O'], format='%Y-%m-%d %H:%M:%S')
synthetic_features
```

```python
synthetic_features['high_season'] = synthetic_features.apply(lambda row: 1 if ((row['Fecha-I'] >= pd.Timestamp(2016,12,15) and row['Fecha-I'] <= pd.Timestamp(2017,3,3)) or 
                                                                               (row['Fecha-I'] >= pd.Timestamp(2017,7,15) and row['Fecha-I'] <= pd.Timestamp(2017,7,31)) or 
                                                                               (row['Fecha-I'] >= pd.Timestamp(2017,9,11) and row['Fecha-I'] <= pd.Timestamp(2017,9,30)) or
                                                                               (row['Fecha-I'] >= pd.Timestamp(2017,12,15) and row['Fecha-I'] <= pd.Timestamp(2018,3,3))) else 0, axis=1)
```

```python
synthetic_features
```

```python
synthetic_features['min_diff'] = synthetic_features.apply(lambda row: (row['Fecha-O'] - row['Fecha-I']).total_seconds() / 60, axis=1)
```

```python
synthetic_features
```

```python
synthetic_features['delay_15'] = synthetic_features.apply(lambda row: 1 if (row['min_diff'] > 15) else 0, axis=1)
```

```python
def check_period_day(date):
    # morning: 5:00 and 11:59 
    if date.hour >= 5 and date.hour < 12:
        return "morning"
    # afternoon (between 12:00 and 18:59)
    elif date.hour >= 12 and date.hour < 19:
        return "afternoon"
    else:
        return "night"
```

```python
synthetic_features['period_day'] = synthetic_features.apply(lambda row: check_period_day(row['Fecha-I']), axis=1)
```

```python
synthetic_features
```

# Prepare Data

```python
synthetic_features.columns
```

```python
# Setear variables categóricas

synthetic_features[['Des-I', 'DIA', 'MES', 'DIANOM', 'TIPOVUELO', 'OPERA', 'high_season', 'period_day']] = synthetic_features[['Des-I', 'DIA', 'MES', 'DIANOM', 'TIPOVUELO', 'OPERA', 'high_season', 'period_day']].astype("category")
```

```python
# Onehotencoder para usar distintos modelos binarios

# Labelizer

# Seleccionamos las columnas que queremos codificar
columnas_a_codificar = ['Des-I', 'DIA', 'MES', 'DIANOM', 'TIPOVUELO', 'OPERA', 'high_season', 'period_day']

# Aplicamos OneHotEncoding usando get_dummies
df_codificado = pd.get_dummies(synthetic_features, columns=columnas_a_codificar, prefix_sep="__")

# Mostramos el DataFrame codificado
df_codificado.columns
```

```python
# Variables

X = df_codificado.drop(columns=['delay_15'])
y = synthetic_features[['delay_15']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, 
```

### Es importante mencionar que existe un desbalanceo de clases, sin embargo, se ejecutarán los modelos y se comparán los F-1 Score para tener una mejor aproximación sobre que modelos tienen mejor desempeño, para posteriormente hacer un tuneo de estos

# Tunning Best Model

### La mayoría de los modelos dan resultados similares, sin embargo puede probar con submuestreo para ver como es el comportamiento.

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler()
X_resampled, y_resampled = rus.fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train_resampled, X_test_resampled, y_train_resampled, y_test_resampled = train_test_split(X_resampled, y_resampled, random_state=42, stratify=y_resampled)

X_train_resampled.shape, y_train_resampled.shape, X_test_resampled.shape, y_test_resampled.shape, 
```

```python
from sklearn.tree import DecisionTreeClassifier

model_DecisionTree = DecisionTreeClassifier()
model_DecisionTree.fit(X_train_resampled, y_train_resampled)

from sklearn.metrics import accuracy_score

y_pred_resampled = model_DecisionTree.predict(X_test_resampled)
accuracy = accuracy_score(y_test_resampled, y_pred_resampled)

print(f"Accuracy: {accuracy}")
print(classification_report(y_test_resampled, y_pred_resampled))
```

### Importancia de las variables para el modelo de Árboles de Decisión con submuestreo

```python
importance = model_DecisionTree.feature_importances_

dict_importance = dict()

for i,v in enumerate(importance):
    dict_importance[X_train.columns[i]] = v

new_importance = dict()

for key in dict_importance.keys():
    if key.split("__")[0] in new_importance.keys():
        new_importance[key.split("__")[0]] += dict_importance[key]
    else:
        new_importance[key.split("__")[0]] = dict_importance[key]

new_importance
```

### Otra técnica es combination sampling, el cual combina el sobremuestro y submuestro

```python
from imblearn.combine import SMOTEENN

smote_enn = SMOTEENN()
X_resampled, y_resampled = smote_enn.fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train_resampled, X_test_resampled, y_train_resampled, y_test_resampled = train_test_split(X_resampled, y_resampled, random_state=42, stratify=y_resampled)

X_train_resampled.shape, y_train_resampled.shape, X_test_resampled.shape, y_test_resampled.shape, 
```

### Accuracy y Matriz de confusión con las clases balanecadas

```python
from sklearn.tree import DecisionTreeClassifier

model_DecisionTree = DecisionTreeClassifier()
model_DecisionTree.fit(X_train_resampled, y_train_resampled)

from sklearn.metrics import accuracy_score

y_pred_resampled = model_DecisionTree.predict(X_test_resampled)
accuracy = accuracy_score(y_test_resampled, y_pred_resampled)

print(f"Accuracy: {accuracy}")
print(classification_report(y_test_resampled, y_pred_resampled))
```

### Importancia de las variables

```python
importance = model_DecisionTree.feature_importances_

dict_importance = dict()

for i,v in enumerate(importance):
    dict_importance[X_train.columns[i]] = v

new_importance = dict()

for key in dict_importance.keys():
    if key.split("__")[0] in new_importance.keys():
        new_importance[key.split("__")[0]] += dict_importance[key]
    else:
        new_importance[key.split("__")[0]] = dict_importance[key]

new_importance
```
